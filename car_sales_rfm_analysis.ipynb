<h2>What is RFM?</h2>
<p>RFM stands for Recency, Frequency, and Monetary, a model used to analyze and segment customers based on their purchasing behavior:</p>
<ul>
    <li><strong>Recency (R):</strong> How recently a customer made a purchase.</li>
    <li><strong>Frequency (F):</strong> How often a customer makes a purchase.</li>
    <li><strong>Monetary (M):</strong> How much money a customer spends on purchases.</li>
</ul>



    
<h2>How to Determine RFM Scores?</h2>
<p>Determining RFM scores involves the following steps:</p>

<h3>Data Collection</h3>
    <p>Collect comprehensive data on customer transactions, including purchase dates, the frequency of purchases, and the total spending amount.</p>
    <h3>Score Calculation</h3>
    <p>Calculate scores for Recency, Frequency, and Monetary value. Typically, customers are segmented into quintiles, with each quintile receiving a score from 1 to 5 (where 1 represents the lowest value and 5 the highest).</p>

<h3>Segmentation</h3>
    <p>Combine the individual RFM scores to generate a composite RFM score for each customer. This enables effective segmentation based on purchasing behavior.</p>

<h3>Analysis</h3>
    <p>Analyze the segmented data to uncover key customer segments, trends, and patterns. This analysis provides valuable insights for developing targeted marketing strategies, enhancing customer retention, and driving sales growth.</p>

<h1 style='color: #0d582b;'>RFM Score Calculations</h1>

<h3>R,F,M Score Data Preparation</h3>

```python
# Recency score
recency_df = df.groupby('CUSTOMERNAME')['DAYS_SINCE_LASTORDER'].min().reset_index()
recency_df.columns = ['CUSTOMERNAME','Recency']

# Frequency Score
frequency_df = df.groupby('CUSTOMERNAME').agg({'ORDERNUMBER':'nunique'})
frequency_df.columns = ['Frequency']

# Monetary Score
monetary_df = df.groupby('CUSTOMERNAME').agg({'SALES':'sum'})
monetary_df.columns = ['Monetary']
```

```python
# Making a new df for RFM calculations.
rfm_df = pd.concat([recency_df.set_index('CUSTOMERNAME'), frequency_df, monetary_df], axis=1)

rfm_df.head()
```

```python
# RFM Score Calculating

rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 5, labels = [5, 4 , 3, 2, 1])
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), 5, labels = [1, 2 , 3, 4, 5])
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], 5, labels = [1, 2 , 3, 4, 5])

rfm_df['RFM_Score'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)

rfm_df.head()
```

<h1 style='color: #0d582b;'>Customer Segmentation By RFM Score</h1>

```python
# Segmentation

segmentation_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}

rfm_df['Segment'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str)
rfm_df['Segment'] = rfm_df['Segment'].replace(segmentation_map, regex=True)
rfm_df
```
